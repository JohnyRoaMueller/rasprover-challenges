```python
# Project based on the RaspRover image UGV_S10_240518.zip -> https://drive.google.com/file/d/1ELeIAsUIQ6ydEsc19Vssc6X0nz1myuJ_/view
# Modified and extended by Johny Roa MÃ¼ller, 2025

# Controller function

##Test whole ctrl
import traceback
import os
import pygame
import time
from base_ctrl import BaseController
import audio_ctrl
from audio_ctrl import play_file
from time import sleep


ctrl =  BaseController('/dev/ttyAMA0', 115200)

# Initialize PyGame
pygame.init()
pygame.joystick.init()

# Check if joystick is connected
if pygame.joystick.get_count() == 0:
    print("No joystick detected")
    exit()

joystick = pygame.joystick.Joystick(0)  # /dev/input/js0
joystick.init()

ctrl_mode_flag = True
ctrl.send_command({"T":3,"lineNum":0,"Text":"----ARCADE---MODE----"})

try:
    while True:
        pygame.event.pump()  # Process events
        

        button_LB = joystick.get_button(6)
        button_RB = joystick.get_button(7)  
        button_LT = joystick.get_button(8)
        button_RT = joystick.get_button(9)          


        if button_LB and button_RB and button_LT and button_RT:
            sleep(0.3)
            ctrl_mode_flag = not ctrl_mode_flag
            if ctrl_mode_flag:
                play_file("control_modes/arcade_mode.mp3")
                print("arcade sound")
                ctrl.send_command({"T":3,"lineNum":0,"Text":"----ARCADE---MODE----"})
            else:
                play_file("control_modes/tank_mode.mp3")
                print("tank sound")
                ctrl.send_command({"T":3,"lineNum":0,"Text":"----TANK---MODE----"})
    
            print(f'ctrl_mode_flag now {ctrl_mode_flag}')

        #-> Arcade Drive
        if ctrl_mode_flag:
            y_axis = -joystick.get_axis(1)  # forward/backwards
            x_axis = joystick.get_axis(0)   # left/right
            left_speed = y_axis + x_axis
            right_speed = y_axis - x_axis

        #-> Tank Drive        
        if not ctrl_mode_flag:
            left_speed = joystick.get_axis(1)  # both left wheels
            right_speed = joystick.get_axis(3) # both right wheels

        
        ctrl.base_speed_ctrl(left_speed, right_speed)


        
        button_x = joystick.get_button(0)
        if button_x:
            play_file("lasers/lasershot_long.mp3")
            print("piu")
            sleep(0.3)


        
        time.sleep(0.05)


except KeyboardInterrupt:
    print("Exiting...")
except Exception:
    traceback.print_exc()
```
